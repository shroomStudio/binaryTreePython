# Binary Tree

##Introducción: En este documento se presenta el código y explicación del método de ordenamiento y busqueda denominado binary tree, el cual es sin duda uno de los algoritmos mas utilzados para el tratamiento de datos que necesitan ser procesados con una determinada jerarquía, uno de los ejemplos mas practicos para poder enter el principo de funcionamiento de este programa es visualizarlo de la siguiente manera: 
Suponga que se tiene que realizar la busqueda de un nombre en una guía telefonica impresa, la forma tradicional es ir buscando a través del ordenamiento alfabetico que la guía tiene por defecto, sin embargo, tambien se puede realizar una busquea binaria: en la primera iteración de la busqueda se abre la guía exactamente por la mitad y se determina si el nombre buscado esta a la izquierda o derecha, suponga se encuentra a la izquierda a la segunda iteración la parte izquierda seleccionada se vuele a abrir exactamente por la mitad y se repite el proceso anterior, y de esa manera hasta llegar al nombre que se busca. 
Bajo el mismo principio es que se implemente el codigo que aquí se muestra. 

```python
class Node: #se declara la clase llamada Node (nodo)
    def __init__(self, data, parent):#se declaran las variables de entrada 
        self.data = data#se declaran las instancias de la clase
        self.parent = parent#data, parent, right_node, left_node
        self.right_node = None
        self.left_node = None


class BinarySearchTree:#Se declara a clase BinarySearchTree
    def __init__(self):#Se declaran las variables de entrada 
        self.root = None#se declara la instancia llamada root
        
    def insert(self,data):#se define la función de la instancia insert
        if self.root is None:#sentencia condicional que verifica si existe la raiz
            self.root = Node (data,None)#en dado caso que no exista asigna el dato ingresado
            #como raiz
        else:#Si la raiz no esta vacia 
            self.insert_node(data,self.root)#al no estar vacia manda a llamamar a la instancia 
            #insert_node
    
    def insert_node(self,data,node):#se define la instancia insert_node
        if data< node.data:#sentencia condicional que verifica si el dato es menor que el nodo 
            if node.left_node:#de referencia, condicinal que verifica si el nodo izquierdo existe
                self.insert_node(data, node.left_node)#si no existe inserta el nuevo dato en el lugar
            else:#en dado caso que exista un nodo del lado izquierdo 
                node.left_node = Node(data,node)#vuelve a llamar la instancia nodo para realizar una nueva
                #comparacación
        else:#en dado caso que el nuevo dato es mayor que el nodo existe
            if node.right_node:#verifica si del lado derecho existe un nodo
                self.insert_node(data, node.right_node)#si no existe inserta el nuevo dato
            else:#del caso contrario en que exista un nodo
                node.right_node= Node(data, node)#vuelve a llamar la instancia nodo y realiza una nueva
                #comparación.
    def traverse (self):#se define la instancia traverse que de manera general imprime el arbol 
        #de menor a mayor.
        if self.root is not None:#sentencia comparita que evalua si existe raiz del arbol
            self.traverse_in_order(self.root)#en dado caso que existe la instancia se vuelve a
            #llamar así misma. 
    def traverse_in_order(self,node):#se define la instancia traverse_in_order
        if node.left_node:#sentencia condicional que evalua si existe nodo al lado izquiero
            self.traverse_in_order(node.left_node)#en dado caso que existe se vuelve a llamar 
            #la instancia para ver si existe un dato mas anidado del lado izquierdo
        print(node.data)#imprime el dato mas anidado 
    
        if node.right_node:#sentencia condicional que evalua si existen datos de el lado derecho
            self.traverse_in_order(node.right_node)#en dado caso que exista vuelve a llamar la instancia
            #para buscar el dato mas anidado a la derecha
    def maximo(self):#se define la instancia maxima
        if self.root:#sentencia condicional que evalua si existe una raiz
            return self.max_tree(self.root)#si existe devuelve el valor maximo llamando a la instancia 
            #max.tree
    def max_tree(self,node):#se define la instancia max_tree
        if node.right_node:#setencia condicional que evalua si existe nodo a la derecha
            return self.max_tree(node.right_node)#si existe manda a llamar a la instancia max_tree
        return node.data#en caso que no exista el nuevo valor es el maximo y es devuelto como tal
    def minimo(self):#se define al instancia minimo
        if self.root:#verifica si existe raiz
            return self.min_tree(self.root)#en dado que exista manda a llamar la instancia 
            #min_tree
    
    def min_tree(self,node):#se define la instancia min_tree
        if node.left_node:#verifica si existe nodo del lado izquierdo
            return self.min_tree(node.left_node)#en dado caso que exista manda a llamar 
            #la instancia min_tree nuevamente
        return node.data#en dado que no exista define el nuevo dato como el minimo
    
    
    def remove(self,data):#se define la instania remove
        if self.root:#se verifica si existe una raiz
            self.remove_node(data,self.root)#en dado caso que exista raiz manda a llamar
            #a la instancia remove_node
    
    def remove_node(self,data,node):#se define la instancia remove_node
        
        if node is None:#si el nodo esta vacio
            return#sale de la instancia ya que no hay nada que remover
        
        if data < node.data:#sentencia condicional que evalua sí el dato es menor que el 
            #nodo existente
            self.remove_node(data,node.left_node)#en caso de que así sea se llama nuevamente la 
            #instancia para que verifique nuevamente si existe uno menor
        elif data > node.data:#evalua si el dato es mayor ue el nodo existente
            self.remove_node(data,node.right_node)#en dado caso que si existe llama nuevamente 
            #la instancia nuevamente
        else:#en dado caso que el dato ingresado no es una hoja se procede a realizar los casos
            if node.left_node is None and node.right_node is None:##sentencia condicional que evalua sí
                #si el nodo izquierdo esta vacio al igual que el derecho
                print("Removing a leaf node..%d"%node.data)#se comienza a remover una hoja 
                #ya que se verifico que no tiene hijos
                
                parent = node.parent#se define el padre como igual a la instancia node.parent
                
                if parent is not None and parent.left_node == node:##sentencia condicional que evalua sí
                    #si el padre tiene dado y y si el hijo del lado izquierdo es igual al nodo
                    parent.left_node = None#en dado caso que así sea el nodo izquierdo se borra
                if parent is not None and parent.right_node == node:#sentencia condicional que evalua sí
                    #si el padre tienen y es igual al dato ingresado
                    parent.right_node = None#en dado caso que se cumple se remueve el nodo derecho
                    
                if parent is None:#si el padre está limpio
                    self.root = None#la raiz es removida 
                
                del node#devuelve el nodo
            elif node.left_node is None and node.right_node is not None:#sentencia condicional que evalua sí
                #el nodo izquierdo está vacio y el nodo derecho no lo está
                print("Removing a node with single right child")#en dado que se cumpla el nodo solo tiene una
                #rama al lado derecho y se procede a remover y reorganizar el arbol
                
                parent = node.parent#se define el padre como igual a la instancia node.parent
                
                if parent is not None:#si el nodo continiene información
                    if parent.left_node == node:#se verifica si el nodo izquierdo es igual al dato
                        parent.left_node = node.left_node#de ser así se remueve el nodo izquierdo
                    if parent.rightChild == node:#se verifica si el nodo derecho es igua al dato
                        parent.right_node = node.right_node#de ser así se remueve el nodo derecho
                else:#en dado caso que el padre no contiene información
                    self.root = node.right_node#se define la raiz como igual al nodo derecho
                node.right_node.parent = parent#se define el nodo derecho como el padre
                del node#se regresa el dato
            elif node.right_node is None and node.left_node is not None:#sentencia condicional que evalua sí
                #el nodo derecho está vacio y el nodo izquierdo no lo está
                print("Removing a node with single left child")#en dado que se cumpla el nodo solo tiene una
                #rama al lado izquierda y se procede a remover y reorganizar el arbol
                
                parent = node.parent#se define el padre como igual a la instancia node.parent
                
                if parent is not None:#si el nodo continiene información
                    if parent.left_node == node:#se verifica si el nodo izquierdo es igual al dato
                        parent.left_node = node.left_node#se define al padre como igual al nodo izquierdo
                    if parent.right_node == node:#se verifica si el nodo derecho es igual al dato
                        parent.right_node = node.left_node#se define al nodo derecho como igual al izquierdo
                else:#en dado caso que el padre no contiene información
                    self.root = node.left_node#se define la raiz como igual al nodo izquerdo
                node.left_node.parent = parent
                del node
            
            else:#del caso que el dato no cae en ninguna de las sentencias anteriores 
                print("Removing node with two children")#se define que tiene dos hijos
                predecessor = self.get_predecessor(node.left_node)#se define el predecessor
                #como igual a la instancia get_precedessor
                
                temp = predecessor.data#se crea una variable temporal para almacenar el dato anterior 
                predecessor.data = node.data#se define la variable predecessor.data como igual
                #al nuevo dato
                node.data=temp#y el nuevo dato pasa a ser el nodo anterior guardado con el fin de conservarlo
                
                self.remove_node(data,predecessor)#se llama la instancia remove_node
    def get_predecessor(self,node):#se define la instancia get_predecessor
        if node.right_node:#se evalua si existe dato a la derecha
            return self.get_predecessor(node.right_node)#si existe se manda a llamar la instancia 
            #get_predecessor nuevamente
        return node#regresa el nodo nuevo
        
    
```
